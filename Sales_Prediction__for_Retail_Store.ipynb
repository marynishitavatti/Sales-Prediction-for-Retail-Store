```python
# Feature Engineering
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
```

```python
df.head()
```

```python
df.drop("Date", axis=1, inplace=True)
```

```python
df.columns
```

```python
df["StateHoliday"].unique()
```

```python
df["StateHoliday"] = df["StateHoliday"].astype(str)
encoder = LabelEncoder()
df["StateHoliday"] = encoder.fit_transform(df["StateHoliday"])
df["StateHoliday"].unique()
```

```python
# Keep only days when the store was open
df = df[df["Open"] == 1]
```

```python
df.shape
```

```python
# Feature Selection
X = df[["Store","DayOfWeek","Promo","StateHoliday","SchoolHoliday","Year","Month","Day"]]
y = df["Sales"]
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,shuffle=False)
```

```python
# Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict Sales
predictions = model.predict(X_test)
```

```python
# Evaluate Model
print("MAE :", mean_absolute_error(y_test, predictions))
print("RMSE :", np.sqrt(mean_squared_error(y_test, predictions)))
print("R² Score :", r2_score(y_test, predictions))
```

```python
# Predict Future Sales (User Input)
store = int(input("Enter Store ID: "))
day = int(input("Enter Day of Week (1-7): "))
promo = int(input("Promo (1=Yes, 0=No): "))
state = int(input("State Holiday (0-3): "))  # 0-No State Holiday, 1-Public Holiday, 2-Easter Holiday, 3-Christmas
school = int(input("School Holiday (1=Yes, 0=No): "))
year = int(input("Year: "))
month = int(input("Month: "))
date = int(input("Day: "))

future_data = pd.DataFrame([[store, day, promo, state, school, year, month, date]],columns=X.columns)
prediction = model.predict(future_data)
print(f"\nPredicted Future Sales: ₹{prediction[0]:.2f}")
```
